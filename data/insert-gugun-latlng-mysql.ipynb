```python
import pymysql
import pandas as pd

pd.set_option("display.max_rows", None)
```

```python
db = pymysql.connect(host="aws.connect.psdb.cloud", port=3306, user="", passwd="", db="odeseoul", charset="utf8mb4", ssl={"rejectUnauthorized": True})
```

```python
df_gugun = pd.read_csv("input/자치구.csv")
df_gugun
```

```python
for idx, gugun in df_gugun.iterrows():
  with db.cursor() as cur:
    sql = """
    UPDATE
      location
    SET
      is_seoul_gugun = 1,
      latitude = {},
      longitude = {}
    WHERE
      code = {}
    """ .format(gugun["Y"], gugun["X"], gugun["행정동코드"])

    print(sql)
    cur.execute(sql)
```

```python
db.commit()
db.close()
```
