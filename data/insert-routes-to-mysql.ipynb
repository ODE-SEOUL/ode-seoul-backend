```python
import pymysql
import json
```

```python
db = pymysql.connect(host="aws.connect.psdb.cloud", port=3306, user="", passwd="", db="odeseoul", charset="utf8mb4", ssl={"rejectUnauthorized": True})
```

```python
with open("input/routes.json", "r") as f:
  routes = json.load(f)

linestrings_of_id = dict()

for route in routes:
 linestrings = linestrings_of_id.get(route["id"], [])
 linestrings.append("(" + ",".join(map(lambda point: "{} {}".format(point[1], point[0]), route["coordinates"])) + ")")
 linestrings_of_id[route["id"]] = linestrings

for route in routes:
  with db.cursor() as cur:
    sql = """
    UPDATE
      course
    SET
      route = ST_GeomFromText('MULTILINESTRING({})', 4326)
    WHERE
      id = {}
    """ .format(",".join(linestrings_of_id[route["id"]]), route["id"])

    print(sql)
    cur.execute(sql)
```

```python
db.commit()
db.close()
```
